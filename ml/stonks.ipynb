```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
start_date=input("enter the starting date:")
stocks=input("enter yahoo finance code for the stock:")
end_date=input("enter an ending date:")
data=yf.download(stocks,start_date,end_date, auto_adjust=True)
data.head()
data.describe()
data.Close.plot()
plt.ylabel=("{}prices".format(stocks))
plt.title=("{}Price Series".format(stocks))
plt.show()
```
